```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter, ImageEnhance
```

```python
img= cv2.imread(r"directory\Reference photos\big_hat.jpg") #enter path to your image
#cv2.IMREAD_GRAYSCALE
```

```python
plt.imshow(img)
plt.axis(False)
plt.show()
```

```python
plt.imshow(img[:,:,::-1])
plt.axis(False)
plt.show()
```

```python
grey_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #changes image to black and white
```

```python
plt.imshow(grey_img, cmap=plt.get_cmap('gray'))
plt.axis(False)
plt.show()
```

```python
invert_img = cv2.bitwise_not(grey_img) #inverts the image in order to highlight lines of image
```

```python
plt.imshow(invert_img, cmap=plt.get_cmap('gray'))
plt.axis(False)
plt.show()
```

```python
blur_img = cv2.GaussianBlur(invert_img, (11,11), 0) #blurs image just slightly 
```

```python
plt.imshow(blur_img,cmap=plt.get_cmap('gray'))
plt.axis(False)
plt.show()
```

```python
invert_blur = cv2.bitwise_not(blur_img)
```

```python
plt.imshow(invert_blur, cmap=plt.get_cmap('gray'))
plt.axis(False)
plt.show()
```

```python
sketch_img = cv2.divide(grey_img, invert_blur, scale = 256.0)
```

```python
plt.imshow(sketch_img, cmap=plt.get_cmap('gray'))
plt.axis(False)
plt.show()
```

```python
sketch_denoise = cv2.fastNlMeansDenoising(sketch_img,10,7,21)
```

```python
plt.imshow(sketch_denoise, cmap=plt.get_cmap('gray'))
plt.axis(False)
plt.show()
```

```python
cv2.imwrite(r'directory\Reference photos\big_hat_21.jpg',sketch_denoise) #save file we will enhance it in next steps
```

```python
image = Image.open(r'directory\Reference photos\big_hat_21.jpg') #open saved image
im3 = ImageEnhance.Contrast(image)
image_dark = im3.enhance(10.0)
```

```python
plt.imshow(image_dark, cmap=plt.get_cmap('gray'))
plt.axis(False)
plt.show()
```

```python
image_dark.save(r'directory\Reference photos\big_hat_dark.jpg') #save image so we can continue to darken the lines
```

```python
image_dark = cv2.imread(r"directory\Reference photos\big_hat_dark.jpg") #upload image again
sketch_denoise2 = cv2.fastNlMeansDenoising(image_dark,150,21,41)
plt.imshow(sketch_denoise2, cmap=plt.get_cmap('gray'))
plt.axis(False)
plt.show()
```

```python
cv2.imwrite(r'directory\Reference photos\big_hat_sketchdn3.jpg',sketch_denoise2) #finally save the final image once happy
```
